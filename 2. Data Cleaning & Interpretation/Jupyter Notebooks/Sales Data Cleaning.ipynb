```python
df_csv['Order_Date'] = pd.to_datetime(df_csv['Order_Date'], errors='coerce', dayfirst=False)
```

```python
df_csv['Units_Sold'] = df_csv['Units_Sold'].astype('int64') 
```

```python
df_csv['Unit_SellingPrice'] = df_csv['Unit_SellingPrice'].astype('float64') 
```

```python
df_csv['Unit_MakingCost'] = df_csv['Unit_MakingCost'].astype('float64')
```

```python
df_csv.drop_duplicates(inplace=True)
```
